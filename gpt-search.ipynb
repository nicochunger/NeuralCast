```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()

OPENAI_KEY = os.getenv("OPENAI_API_KEY")
```

```python
prompt = "Give me a playlist of [NR_TRACKS] [GENRE_TAG] songs. [GENRE_DESCRIPTION] The songs should be listed in plain text. \
Return each song on a new line with this format: 'Artist - Song Title - Year'. Don't include any list numbering, just the song with the information and format I indicated."
```

```python
import json

with open("genre_prompts.json", "r") as f:
    genre_data = json.load(f)

genre_tag = genre_data["genres"][3]["tag"]
genre_description = genre_data["genres"][3]["description"]
nr_tracks = 10

full_prompt = (
    prompt.replace("[NR_TRACKS]", str(nr_tracks))
    .replace("[GENRE_TAG]", genre_tag)
    .replace("[GENRE_DESCRIPTION]", genre_description)
)
print(full_prompt)
```

```python
def call_gpt(prompt: str, model: str = "gpt-4o") -> str:
    client = openai.OpenAI(api_key=OPENAI_KEY)
    response = client.chat.completions.create(
        model=model, messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
```

```python
def gpt_playlist(prompt: str):
    raw = call_gpt(prompt)
    lines = [l.strip().split(" - ") for l in raw.splitlines() if "-" in l]
    return lines
```

```python
playlist = gpt_playlist(full_prompt)
playlist
```

```python
response = call_gpt(
    "Give me in short paragraphs 5 news stories about the lastest in Argentina's politics.",
    model="gpt-4o-search-preview",
)
```

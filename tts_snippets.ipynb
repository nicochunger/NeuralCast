# TTS snippets of NeuralCast and NeuralForge

```python
from pathlib import Path
from openai_utils import tts, openai_speech
from tqdm.notebook import tqdm
import os
```

## General snippets

```python
# Load NeuralCast host snippets from a text file and strip newlines
with open("NeuralCast/tts_snippets/neuralcast_general.txt", "r") as file:
    neuralcast_snippets = [line.strip() for line in file]
```

```python
for idx, snippet in tqdm(
    enumerate(neuralcast_snippets),
    total=len(neuralcast_snippets),
    desc="Generating TTS snippets",
):
    speech_file_path = Path("NeuralCast/songs/Host Snippets") / f"Snippet_{idx:02}.mp3"

    tts(snippet, speech_file_path)
```

## Melodic House intros

```python
with open("NeuralCast/tts_snippets/chill_melodic_house_intros.txt", "r") as file:
    chill_melodic_house_intros = [line.strip() for line in file]

for idx, snippet in tqdm(
    enumerate(chill_melodic_house_intros),
    total=len(chill_melodic_house_intros),
    desc="Generating Chill Melodic House Intros",
):
    output_path = (
        Path("NeuralCast/songs/intros_chill_melodic_house")
        / f"Chill_Melodic_House_Intro_{idx:02}.mp3"
    )

    tts(snippet, output_path)
```

## Aspen Night intros

```python
with open("NeuralCast/tts_snippets/aspen_night_intros.txt", "r") as file:
    aspen_night_intros = [line.strip() for line in file]

for idx, snippet in tqdm(
    enumerate(aspen_night_intros),
    total=len(aspen_night_intros),
    desc="Generating Aspen Night Intros",
):
    output_path = (
        Path("NeuralCast/songs/intros_aspen_night") / f"Aspen_Night_Intro_{idx:02}.mp3"
    )
    # Create only if it doesn't exist
    if os.path.exists(output_path):
        print(f"File already exists, skipping: {output_path}")
        continue
    else:
        tts(snippet, output_path)
        print(f"Generated speech for Aspen Night intro: {output_path}")
```

# NeuralForge intros

```python
with open("NeuralForge/tts_snippets/neuralforge_general.txt", "r") as file:
    neuralforge_snippets = [line.strip() for line in file]

for idx, snippet in tqdm(
    enumerate(neuralforge_snippets),
    total=len(neuralforge_snippets),
    desc="Generating NeuralForge TTS snippets",
):
    speech_file_path = (
        Path("NeuralForge/songs/Host Snippets") / f"NeuralForge_Snippet_{idx:02}.mp3"
    )

    tts(
        text=snippet,
        outfile=speech_file_path,
    )
```
